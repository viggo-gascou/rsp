```python
import os
import sys
import typing as t

if t.TYPE_CHECKING:
    from types import TracebackType


class HiddenPrints:
    """Context manager which removes all terminal output."""

    def __enter__(self) -> None:
        """Enter the context manager."""
        self._original_stdout = sys.stdout
        self._original_stderr = sys.stderr
        sys.stdout = open(os.devnull, "w")
        sys.stderr = open(os.devnull, "w")

    def __exit__(
        self,
        exc_type: t.Type[BaseException],
        exc_val: BaseException,
        exc_tb: "TracebackType",
    ) -> None:
        """Exit the context manager."""
        sys.stdout.close()
        sys.stderr.close()
        sys.stdout = self._original_stdout
        sys.stderr = self._original_stderr

```

```python
import pandas as pd
import os
import PIL
import logging
from tqdm import tqdm

logging.getLogger("root").setLevel(logging.WARNING)
data_path = "/Users/viggogascou/Downloads/CelebAMask-HQ/CelebA-HQ-img-512/"

img_list = sorted([data_path + path for path in os.listdir(data_path) if path.endswith('.jpg') or path.endswith('.png')])
deteced_aus = pd.DataFrame()
for img in tqdm(img_list[:50]):
    with HiddenPrints():
        mixed_prediction = detector.detect_image(img, face_detection_threshold=0.9)
    print(mixed_prediction["input"].str.replace(data_path, ""))
    deteced_aus = pd.concat([deteced_aus, mixed_prediction])


    
deteced_aus
```

```python
from rsp.utils import image_grid
import os
from rsp.constants import *
from rsp.stateclass import Q
from PIL import Image
import torch
```

```python

aus_= ["AU04", "AU12", "AU06", "AU10"]

path = RESULTS_DIR / "figures" / "convergence_explorer"
os.makedirs(path, exist_ok=True)

aus_umgs = {}
for au in aus_:
    au_dir = path / au
    imgs = []
    for i in range(10, 100, 10):
        img_path = au_dir / f"image_{i}_{au}_edited.png"
        img = Image.open(img_path).convert("RGB")
        imgs.append(img)

    aus_umgs[au] = imgs

final_imgs = []
for au in aus_:
    final_imgs.extend(aus_umgs[au])

grid = image_grid(final_imgs, rows=len(aus_), cols=len(final_imgs) // len(aus_))
grid.save(path / "convergence_explorer.png")
```
